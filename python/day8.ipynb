```python
%%ipytest

logging.getLogger().setLevel(logging.DEBUG)

def check_visibility(dd, rr, cc):
    hh = dd[rr][cc]

    # check to the left
    yy, xx = (rr, cc-1)
    while xx >= 0:
        if dd[yy][xx] >= hh:
            break
        xx -= 1
    else:
        return True

    # check to the right
    yy, xx = (rr, cc+1)
    while xx < len(dd[0]):
        if dd[yy][xx] >= hh:
            break
        xx += 1
    else:
        return True

    # check to the top
    yy, xx = (rr-1, cc)
    while yy >= 0:
        if dd[yy][xx] >= hh:
            break
        yy -= 1
    else:
        return True

    # check to the bottom
    yy, xx = (rr+1, cc)
    while yy < len(dd):
        if dd[yy][xx] >= hh:
            break
        yy += 1
    else:
        return True

    return False

def part1(puzzle_input):
    if not os.path.exists(puzzle_input):
        return

    answer = 0

    with open(puzzle_input) as ff:
        dd = [ [ int(digit) for digit in xx.strip() ] for xx in ff.readlines() ]

    hh = [ [ " " ] * len(dd) for _ in range(len(dd)) ] 
    
    # dd is row-wise (dd[0] is the first row)
    ###############################################################################################################
    
    answer = 0
    for rr in range(len(dd)):
        for cc in range(len(dd[rr])):
            # if tree is on edge
            if rr == 0 or rr == (len(dd) - 1):
                hh[rr][cc] = 'X'
            elif cc == 0 or cc == (len(dd[rr]) - 1):
                hh[rr][cc] = 'X'
            else:
                if check_visibility(dd, rr, cc):
                    hh[rr][cc] = 'X'

    print()
    for rr in range(len(dd)):
        for cc in range(len(dd[rr])):
            if hh[rr][cc] == 'X':
                answer += 1
                print('X', end="")
            else:
                print(".", end="")
        print()
    print()
    
    #pprint.pprint(hh)
    ###############################################################################################################
    
    return answer

def test_part1():
    assert part1(os.path.join("..", "dat", "day8_test.txt")) == 21
    
    
part1(os.path.join("..", "dat", "day8.txt"))
```
