```python
def snafu2dec(snafu):
    dec = 0
    for ii, vv in enumerate(reversed(snafu)):
        xx = 5 ** ii
        if vv == '-':
            vv = -1
        elif vv == '=':
            vv = -2
        else:
            vv = int(vv)

        dec += (vv*xx)

    return dec

def dec2snafu(dec):
    snafu = []
    rem = dec
    ii = 1
    while rem > 0:
        xx = 5 ** ii
        mm = (rem % xx) // (5 ** (ii-1))
        if mm == 4:
            rem += xx
            snafu.insert(0, '-')
        elif mm == 3:
            rem += xx
            snafu.insert(0, '=')
        else:
            rem -= mm
            snafu.insert(0, str(mm))
        rem = (rem // xx) * xx
        ii += 1

    return "".join(snafu)

def part1(puzzle_input):
    dd = read_input(puzzle_input)

    ans = 0
    for ll in dd:
        dec = snafu2dec(ll)
        print(ll, dec)
        ans += dec
    
    return dec2snafu(ans)

def test_part1():
    ans = part1(os.path.join(os.path.join("..", "dat", "day25_test.txt")))
    print(ans)
    assert ans == '2=-1=0'

test_part1()
ans = part1(os.path.join(os.path.join("..", "dat", "day25.txt")))
print(ans) # 35422591760336
```
