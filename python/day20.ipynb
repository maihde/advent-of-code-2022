```python
def read_input(puzzle_input):
    with open(puzzle_input) as ff:
        dd = ff.readlines()

    return [ int(xx.strip()) for xx in dd ]

def test_read_input():
    data = read_input(os.path.join(os.path.join("..", "dat", "day20_test.txt")))
    print(data)

test_read_input()
```

```python
class Item():
    def __init__(self, pos, val):
        self.pos = pos
        self.val = val
        self.prev = None
        self.next = None

def part1(puzzle_input):
    enc_data = read_input(puzzle_input)
    dec_data = {}
    for ii, vv in enumerate(enc_data):
        dec_data[ii] = Item(ii, vv)
        if ii > 0:
            dec_data[ii].prev = dec_data[ii-1]
    dec_data[0].prev = dec_data[len(enc_data)-1]

    head = dec_data[0]

    for ii, vv in enumerate(enc_data):
        if ii < len(enc_data) - 1:
            dec_data[ii].next = dec_data[ii+1]
        else:
            dec_data[ii].next = dec_data[0]

    for ii, vv in enumerate(enc_data):
        item = dec_data[ii]
        #print("Moving %s orignally at %s by %s", vv, item.pos, ii)
        if vv > 0: # shift right
            for _ in range(vv):
                left_tmp = item.prev
                right_tmp = item.next

                item.next = right_tmp.next
                item.next.prev = item
                right_tmp.next = item
                right_tmp.prev = item.prev
                item.prev = right_tmp
                left_tmp.next = right_tmp
                if item == head:
                    head = right_tmp
                elif right_tmp == head:
                    head = item
                
        elif vv < 0: # shift left
            for _ in range(abs(vv)):
                left_tmp = item.prev
                right_tmp = item.next
                
                item.prev = left_tmp.prev
                item.prev.next = item
                left_tmp.prev = item
                left_tmp.next = item.next
                item.next = left_tmp
                right_tmp.prev = left_tmp
                if item == head:
                    head = right_tmp
                elif left_tmp == head:
                    head = item
    
        """
        item = head
        for _ in range(len(enc_data)):
            print(item.val, ",", end="")
            item = item.next
        print("")
        """
        
    zero_pos = None
    vals = []
    for ii in range(len(enc_data)):
        vals.append(item.val)
        if item.val == 0:
            zero_pos = ii
        item = item.next

    #print(vals)
    ans = 0
   
    idx = (zero_pos + 1000) % len(vals)
    print("Pos %s Val %s" % (idx, vals[idx]))
    ans += vals[idx]

    idx = (zero_pos + 2000) % len(vals)
    print("Pos %s Val %s" % (idx, vals[idx]))
    ans += vals[idx]

    idx = (zero_pos + 3000) % len(vals)
    print("Pos %s Val %s" % (idx, vals[idx]))
    ans += vals[idx]

    return ans

def test_part1():
    ans = part1(os.path.join(os.path.join("..", "dat", "day20_test.txt")))
    #print(ans)
    assert ans == 3

test_part1()

ans = part1(os.path.join(os.path.join("..", "dat", "day20.txt")))
print(ans)
    
```
